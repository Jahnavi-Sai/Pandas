# What is GDP?
GDP (Gross Domestic Product) is the total monetary value of all final goods and services produced within a country's borders in a specific time period (usually a year or a quarter).

It is one of the most important indicators used to measure the economic performance of a country.





GDP ``=``total income a country earns from everything it produces and sells in a year — including goods (like cars, food, electronics) and services (like banking, education, tourism).

``Why is GDP Important?``

| Purpose                      | Explanation                                                                 |
| ---------------------------- | --------------------------------------------------------------------------- |
| Measures Economic Health  | A rising GDP means economic growth; a falling GDP may signal a recession.   |
| Compares Living Standards | Higher GDP per capita usually means a higher standard of living.            |
| Global Comparison         | Used to compare economies across countries (e.g., India vs. USA vs. China). |
| Policy Making             | Helps governments make decisions on interest rates, taxes, and spending.    |

# The distplot

A distplot (short for distribution plot) is used to visualize the distribution of a continuous (numeric) variable. 

It shows how values in the data are distributed, often highlighting patterns like normal distribution, skewness, or multimodal behavior.

the distplot is a combination of histogram with a line (density plot ) on it. we used the one univariaent analysis

``A typical dist plot may include:``

Histogram – a bar plot showing how frequently values appear in intervals (bins).

KDE (Kernel Density Estimate) – a smooth curve that estimates the probability density function of the variable.

```python
vis1=sns.distplot(df["InternetUsers"]) 
```

```python
vis2=sns.distplot(df["InternetUsers"],bins=10) # 10 bins=10 total histogram
```

# The box plot:

A box plot is a standardized way of displaying the distribution of a numerical variable based on five summary statistics:

--Minimum (excluding outliers)

--1st Quartile (Q1) – 25th percentile

--Median (Q2) – 50th percentile

--3rd Quartile (Q3) – 75th percentile

--Maximum (excluding outliers)

--t also highlights outliers and skewness.

``Purpose of Box Plot:``

1.To understand data spread and symmetry

2.To compare distributions across groups

3.To identify outliers

4.To analyze central tendency and variability






```python
vis4=sns.boxplot(data=df,x='IncomeGroup',y='BirthRate') 
```

# The LM Plot

An lmplot in Seaborn is a scatter plot with a linear regression line (fit line) and confidence interval. 
It shows the relationship between two continuous variables, with optional categorical separation (like income group, gender, etc)

``Use Cases:``

Visualize linear trends between variables (e.g., GDP vs Population)

Detect correlation strength and direction

Show groupwise regression using the hue argument

Great for exploratory data analysis (EDA)



```python
vis5=sns.lmplot(data=df,x='InternetUsers',y='BirthRate') 
```

if we use fit reg =True then it gives regression line False then it delete regression line

```python
vis6=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True) 
```

```python
vis7=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False) 
```

```python
vis6=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True, hue='IncomeGroup') 
# hue- Color-code the data points and regression lines by each category
# if we want to understand the graph divide into 4 parts
```
